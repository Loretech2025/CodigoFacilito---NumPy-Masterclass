# Entendiendo tipos de datos en Python vs. NumPy

Lo mágico de NumPy y lo que lo hace más rápido que usar solamente Python solito, es que NumPy escribe sus métodos y objetos basados en C, recordando la lección anterior, vimos que las variables en Python eran más que sólo variables, tenían una abstracción de objetos muy grandes y era lo que hacía que Python fuera tan lento, pero NumPy implementa la idea de C y hace que todo lo que escribas sea (en el interior) objetos de C puros, por ejemplo, comparemos una pequeña tabla de ambos tipos de datos.

## Tipos de datos en Python:

En **Python**, los tipos son bastante sencillos, los podemos observar en esta tabla:

| Tipo de Dato   | Ejemplo            | Descripción                                           |
|----------------|--------------------|-------------------------------------------------------|
| `int`          | `5`, `-3`, `42`    | Números enteros, positivos o negativos                |
| `float`        | `3.14`, `-2.5`     | Números decimales, representan números reales         |
| `str`          | `'hola'`, `'123'`  | Cadena de texto, una secuencia de caracteres          |
| `bool`         | `True`, `False`    | Valores booleanos, representan valores de verdad      |
| `list`         | `[1, 2, 3]`        | Colección ordenada de elementos, puede contener cualquier tipo de dato |
| `tuple`        | `(4, 5, 6)`        | Colección ordenada e inmutable de elementos           |
| `dict`         | `{'clave': 'valor'}` | Diccionario, colección de pares clave-valor        |
| `set`          | `{1, 2, 3}`        | Colección desordenada de elementos únicos             |
| `NoneType`     | `None`             | Representa la ausencia de un valor                    |
| `complex`      | `3+4j`, `1.5-2j`   | Números complejos, usados en cálculos científicos     |

## Tipos de datos en NumPy

En NumPy, los tipos de datos son más complejos y tenemos más control sobre el uso de la memoria, esto para optimizar.

| Tipo de Dato       | Ejemplo              | Descripción                                                                                   |
|--------------------|----------------------|-----------------------------------------------------------------------------------------------|
| `int32`            | `np.array([1, 2, 3], dtype=np.int32)` | Entero de 32 bits, ocupa menos espacio que `int64` pero con un rango menor                   |
| `int64`            | `np.array([1, 2, 3], dtype=np.int64)` | Entero de 64 bits, útil para valores enteros grandes                                         |
| `float32`          | `np.array([1.0, 2.5, 3.7], dtype=np.float32)` | Decimal de 32 bits, menos precisión y menor consumo de memoria que `float64`        |
| `float64`          | `np.array([1.0, 2.5, 3.7], dtype=np.float64)` | Decimal de 64 bits, mayor precisión y más usado para cálculos científicos            |
| `complex64`        | `np.array([1+2j, 3+4j], dtype=np.complex64)` | Número complejo de 64 bits, usado para cálculos complejos                                  |
| `complex128`       | `np.array([1+2j, 3+4j], dtype=np.complex128)` | Número complejo de 128 bits, para precisión adicional                                     |
| `bool`             | `np.array([True, False], dtype=np.bool_)` | Valores booleanos (`True` o `False`), típicos en operaciones lógicas                       |
| `str_`             | `np.array(['a', 'b', 'c'], dtype=np.str_)` | Cadenas de texto de longitud fija, usadas en operaciones de texto                          |
| `object`           | `np.array(['a', 1, 3.5], dtype=np.object_)` | Tipo de datos para objetos generales de Python (heterogéneos)                              |
| `datetime64`       | `np.array(['2024-01-01'], dtype=np.datetime64)` | Representación de fechas y horas en formato de 64 bits                            |
| `timedelta64`      | `np.array([5], dtype=np.timedelta64)` | Diferencias de tiempo (intervalos) en formato de 64 bits                                    |
| `uint8`            | `np.array([255], dtype=np.uint8)` | Entero sin signo de 8 bits, usado en imágenes y valores de 0 a 255                          |
| `uint16`           | `np.array([65535], dtype=np.uint16)` | Entero sin signo de 16 bits, valores entre 0 y 65535                                        |
| `uint32`           | `np.array([4294967295], dtype=np.uint32)` | Entero sin signo de 32 bits, para valores más grandes                                     |
| `uint64`           | `np.array([18446744073709551615], dtype=np.uint64)` | Entero sin signo de 64 bits, para valores enteros grandes                          |

## Fundamentos de Arreglos en NumPy

Para aprovechar al máximo el poder de NumPy, es fundamental comprender su concepto clave: **el ndarray** (o "arreglo" para los compas). Los arreglos de NumPy son similares a las listas en Python, pero, a diferencia de las listas, están optimizados para el manejo de grandes cantidades de datos numéricos y tienen características que los hacen mucho más eficientes:

1. **Tamaño fijo desde su creación**: Una vez que se crea un arreglo, su tamaño no cambia, lo que permite a NumPy gestionar la memoria de manera más eficiente.
2. **Tipo de dato homogéneo**: Todos los elementos en un arreglo deben ser del mismo tipo (como enteros o flotantes), lo cual optimiza el almacenamiento y las operaciones matemáticas.
3. **Multidimensionalidad explícita**: Los arreglos de NumPy pueden tener múltiples dimensiones (por ejemplo, matrices 2D o tensores 3D), lo que permite trabajar de manera natural con estructuras de datos complejas.

Además de estas características, los arreglos de NumPy permiten realizar **operaciones vectorizadas** (que veremos más adelante), lo que acelera los cálculos en comparación con las listas de Python y es esencial en aplicaciones de ciencia de datos y aprendizaje automático.

````python
# Creémos nuestros primeros ndarrays o "arreglos"
lista_de_enteros = [20, 4, 9, 6]
arreglo_de_enteros = np.array(lista_de_enteros)

# Lo voy a imprimir
arreglo_de_enteros
````

````python
# Veamos el tipo de este arreglo
arreglo_de_enteros.dtype
````

````python
# El tipo de los arreglos se infiera automáticamente. Pero lo puedes asignar manual en la definición de este
arreglo_de_flotantes = np.array([3.15, 20, 30], dtype = 'float64')

# Imprimamos el arrelgo y el tipo en la misma linea
arreglo_de_flotantes, arreglo_de_flotantes.dtype
````

````python
# Veamos la 'forma' de este arreglo
arreglo_de_flotantes.shape
````

````python
# Definamos un arreglo en dos dimensiones
arreglo_2d = np.array([[1,2, 3], [4, 5, 6]], dtype = 'int64')
arreglo_2d
````

# Creando Arrays más fuertes
Especialmente para arreglos más grandes, es mucho más eficiente crear arreglos con métodos y funciones ya hechas en NumPy, algunos ejemplos son:

````python
# Crear un arreglo de 0's de tipo entero
np.zeros(100, dtype=int)
````

````python
# Crear un arreglo de tipo flotante de 4x3 lleno de unos
np.ones((4, 3), dtype=float)
````

````python
# Crear un arreglo de 2x10 lleno de 8.76
np.full((2, 10), 8.76)
````

````python
# Crear arreglo con una secuencia lineal empezando en 0 y terminando en 40
# (como lo que pasa cuando haces range(0, 40, 2) en Python normalito)
np.arange(0, 40, 2)
````

````python
# Crear un arreglo de 5 valores espaciados equidistantemente entre 0 y 1
np.linspace(0, 1, 5)
````

````python
# Aleatorios del 0 al 1 de una forma 4x5!
np.random.random((4, 5))
````

````python
# Aleatorio enteros del 3 al 20 con forma de 1x3
np.random.randint(3, 20, size = (1, 3))
````
